'หุ้นน้ำมันยังไงดี' ประกอบไปด้วยการวิเคราะห์ทั้งหมด 4 ส่วน **" เพื่อประกอบกันเป็นส่วนช่วยในการตัดสินใจลงทุนในหุ้นน้ำมันของบุคคลในอนาคต "** ได้แก่

1. **การวิเคราะห์อัตราส่วนทางการเงินพื้นฐาน** เพื่อเข้าใจศักยภาพทางการเงินเบื้องต้นของบริษัท
2. **การวิเคราะห์การเปลี่ยนแปลงของราคาน้ำมันดิบโลก ราคาหุ้นน้ำมัน และค่าเงินดอลล่าสหรัฐ** เพื่อเข้าใจภาพรวมการเปลียนแปลงในระยะเวลาย้อนหลัง 5 ปี (2018-2022)
3. **การวิเคราะห์ค่าความสัมพันธ์ (correlation) ระหว่างราคาน้ำมันดิบกับราคาหุ้นน้ำมันแต่ละตัวและค่าเงินดอลล่าสหรัฐ** เพื่อเข้าใจลักษณะการแปรผันของปัจจัยเหล่านี้ต่อราคาหุ้นน้ำมันว่ามีความสัมพันธ์กันมากน้อยเพียงใด
4. **เปรียบเทียบวิธีการลงทุนแบบ lump sum** (การลงทุนด้วยเงินก้อนใหญ่) **และการลงทุนแบบ DCA** หรือ Dollar-Cost Averaging ที่แบ่งทะยอยการลงทุนอย่างสม่ำเสมอ เพื่อช่วยประกอบในการตัดสินใจลงทุนในหุ้นน้ำมันในอนาคต

หมายเหตุ* การพิจารณาข้อมูลเชิงปริมาณย้อนหลังเป็นเพียงส่วนหนึ่งในการใช้การประกอบการตัดสินใจการเลือกซื้อหุ้นน้ำมัน การพิจารณาข้อมูลเพิ่มเติมจากเอกสารรายงานประจำปีและข้อมูลข่าวสารเกี่ยวกับบริษัทเป็นอีกปัจจัยสำคัญในการตัดสินใจ

***Insight 1.การวิเคราะห์อัตราส่วนทางการเงินพื้นฐาน*** ในการศึกษานี้จะใช้ EPS P/E ์Net Profit Margin และ D/E ในการพิจารณา โดยใช้ข้อมูลปี 2562 - 2565

**Net Profit Margin** เป็นอัตราส่วนทางการเงินที่เป็นการเปรียบเทียบระหว่าง “กำไรสุทธิ (Net Profit)” เทียบกับ “ยอดขาย (Sales)” เป็นอัตราส่วนที่บอกถึงความสามารถในการทำกำไรของบริษัทได้เป็นอย่างดี Net Profit Margin มีค่าสูงแสดงว่าสินค้าและบริการของบริษัทนั้นมีความสามารถในการทำกำไรที่ดี

**EPS (Earning per Shares)** คำนวณมาจาก Net Profit Margin (กำไรสุทธิ) หารด้วยจำนวนหุ้น บ่งบอกถึงความสามารถในการทำกำไรสุทธิต่อหุ้นซึ่งยิ่งมากยิ่งดี

**P/E** คำนวณมาจาก ราคาหุ้นหารด้วยกำไรสุทธิต่อหุ้น แสดงให้เห็นว่าราคาที่ต้องจ่ายต่อกำไรสุทธิทุก 1 บาท ยิ่งน้อยยิ่งดี

**D/E** คำนวณจากหนี้สินรวมหารด้วยส่วนของผู้ถือหุ้น ค่าที่สูงแปลว่าบริษัทมีภาระหนี้ที่สูง

```python
df_fin = pd.read_csv('https://raw.githubusercontent.com/isareehan/oil_invest/main/financial_ratio.csv')
```

```python
#2565_study
df2565 = df_fin.loc[ (df_fin['year']== 2565),:]
df2565.sort_values( ['EPS','P/E','Net Profit Margin','D/E'], ascending=[False,True,False,True] )
```

```python
#2564_study
df2564 = df_fin.loc[ (df_fin['year']== 2564),:]
df2564.sort_values( ['EPS','P/E','Net Profit Margin','D/E'], ascending=[False,True,False,True] )
```

```python
#2563_study
df2563 = df_fin.loc[ (df_fin['year']== 2563),:]
df2563.sort_values( ['EPS','P/E','Net Profit Margin','D/E'], ascending=[False,True,False,True] )
```

```python
#2562_study
df2562 = df_fin.loc[ (df_fin['year']== 2562),:]
df2562.sort_values( ['EPS','P/E','Net Profit Margin','D/E'], ascending=[False,True,False,True] )
```

```python
fin_ratio = df_fin.loc[ ((df_fin['company'] == 'PTTEP')| (df_fin['company'] == 'BCP')| (df_fin['company'] == 'TOP')),:]
fin_ratio['year'] = fin_ratio['year'].astype(str)
```

จากการนำอัตราส่วนทางการเงินพื้นฐาน 4 ปี ย้อนหลังมาจัดเรียงกล่าวคือ **EPS ยิ่งมากยิ่งดี, P/E ยิ่งน้อยยิ่งดี, Net Profit Margin ยิ่งมากยิ่งดี และ D/E ratio ยิ่งน้อยยิ่งดี** มาจัดเรียงตามลำดับ ผู้จัดทำได้เลือกหุ้นที่อยู่อันดับบนอย่างสม่ำเสมอมาทั้งหมด 3 ตัว ได้แก่ **TOP PTTEP และ BCP** มาแสดงเป็นกราฟ เพื่อดูแนวโน้มและ**พิจารณาเปรียบเทียบอัตราส่วนทางการเงิน**แต่ละชนิดของหุ้นแต่ละตัว

Net Profit Margin:  
**PTTEP มีค่าสูงที่สุด**อย่างต่อเนื่อง**แสดงถึงความสามารถในการทำกำไรได้ดี**เมื่อเทียบกับหุ้นของบริษัทอื่น

```python
sns.lineplot(x= 'year', y='Net Profit Margin', data=fin_ratio, hue = 'company')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Earning per Share:  
 บ่งบอกถึงความสามารถในการทำกำไรสุทธิต่อหุ้นซึ่งยิ่งมากยิ่งดี เมื่อเปรียบเทียบภาพรวมจากข้อมูลย้อนหลัง **PTTEP มี Earning per Share ที่ดีที่สุด**

```python
sns.lineplot(x= 'year', y='EPS', data=fin_ratio, hue = 'company')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

P/E:  
ราคาหุ้นหารด้วยกำไรสุทธิต่อหุ้น ยิ่งน้อยยิ่งดี จากกราฟ **BCP** และ **TOP** มีค่า P/E ที่ต่ำสุด แสดงให้เห็นว่า**มีราคาหุ้นเมื่อเทียบกับกำไรแล้วถูกกว่า PTTEP**

```python
sns.lineplot(x= 'year', y='P/E', data=fin_ratio, hue = 'company')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

D/E:  
คือภาระหนี้สินรวมของบริษัทเมื่อเทียบกับส่วนของผู้ถือหุ้น **PTTEP มีค่าต่ำสุด แสดงว่ามีอัตราภาระหนี้ที่ต่ำกว่า**อีกสองบริษัท

```python
sns.lineplot(x= 'year', y='D/E', data=fin_ratio, hue = 'company')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

***Insight 2. การวิเคราะห์การเปลี่ยนแปลงของราคาน้ำมันดิบตลาดโลก (WTI) ราคาหุ้นน้ำมัน

ในส่วนนี้จะมีการรวบรวมข้อมูลและวิเคราะห์ข้อมูลราคาน้ำมันดิบตลาดโลก ราคาหุ้นน้ำมัน และราคาค่าเงินดอลล่าสหรัฐในระยะเวลาย้อนหลัง 5 ปี (2018-2022) **เพื่อเข้าใจภาพรวมและทิศทางการเปลียนแปลงของราคา**เหล่านี้ว่ามีแนวโน้มหรือทิศทางเป็นเช่นไรในระยะเวลาย้อนหลัง 5 ปีที่ผ่านมา

จากกราฟด้านบนพบว่า ราคาหุ้นน้ำมันส่วนใหญ่มีการเปลี่ยนแปลงไปในทิศทางเดียวกัน ยกเว้นหุ้น PTTEP ที่ในช่วงปลายปี 2021 จนถึงสิ้นปี 2022 มีการเปลี่ยนแปลงในลักษณะที่แตกต่างออกไปจากหุ้นน้ำมันตัวอื่นๆ โดยมีราคาเพิ่มสูงขึ้นอย่างมีนัยสำคัญต่างจากหุ้นน้ำมันอื่นๆในช่วงเวลาดังกล่าว

```python
#Plot Upstream
plt.figure(figsize=(8,4), dpi=150)
all_data.plot( kind='line', x='Date', y=['BCP','PTT', 'PTTEP'],
             linestyle='solid', ax=plt.gca(), ylabel='Price (USD)' )
all_data.plot( kind='line', x='Date', y=['Crude_Oil'],
             linestyle='dashed', ax=plt.gca(), ylabel='Price (USD)' )
plt.title('Historical Thailand Upstream Oil Company Stocks and Crude Oil Price (2018-2022)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

จากกราฟด้านบนพบว่า เมื่อทำการวิเคราะห์เฉพาะ **หุ้นน้ำมันกลุ่มต้นน้ำ** (Upstream) ซึ่งเป็นกลุ่มหุ้นที่ทำธุรกิจเกี่ยวข้องกับการสำรวจและผลิตน้ำมันดิบและก๊าซธรรมชาติ พบว่าในช่วงปี 2018 ถึงต้นปี 2021 ราคาหุ้นน้ำมันกลุ่มต้นน้ำมีการเปลี่ยนแปลงไปในทิศทางเดียวกัน แต่ภายหลังจากต้นปี 2021 **หุ้น PTTEP ที่ทำธุรกิจเกี่ยวกับการสำรวจและผลิตเป็นหลัก** มีการเปลี่ยนแปลงลักษณะที่แตกต่างออกไปจากหุ้นน้ำมันตัวอื่นๆ ซึ่ง**การเปลี่ยนแปลงมีลักษณะสอดคล้องกับราคาน้ำมันดิบตลาดโลก** (WTI)

```python
#Plot Midstream
plt.figure(figsize=(8,4), dpi=150)
all_data.plot( kind='line', x='Date', y=['ESSO', 'IRPC', 'PTTGC', 'SRPC', 'TOP'],
             linestyle='solid', ax=plt.gca(), ylabel='Price (USD)' )
all_data.plot( kind='line', x='Date', y=['Crude_Oil'],
             linestyle='dashed', ax=plt.gca(), ylabel='Price (USD)' )
plt.title('Historical Thailand Midstream Oil Company Stocks and Crude Oil Price (2018-2022)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

***Insight 4. เปรียบเทียบการลงทุนแบบ Lump sum และ DCA***

เพื่อช่วยประกอบในการตัดสินใจลงทุนในหุ้นน้ำมันของบุคคลในอนาคต

```python
#choose only PTTEP
asset_prep = all_data.loc[:, ['Date','PTTEP']]
asset = asset_prep.set_index(pd.DatetimeIndex(asset_prep['Date'].values))
```

```python
#calculate & show lump sum investment strategy
adj_close_prices = asset['PTTEP']

#choose the initial investment amount
initial_investment = 240000

#get the last price
stock_latest_value = adj_close_prices[-1]

#get number of share bought on each date with the hypothesis amount (aka lump sum)
num_shares_bought = initial_investment/adj_close_prices

#calculate how much money you would have according to the last price in this data set for investment amount (aka lump sum)
#investing a lump sum of a given date in the past
lump_sum = stock_latest_value*num_shares_bought

```

```python
#create a function to calculate the DCA Strategy
def DCA(start_date, initial_investment):
    #get the investment periods which will be 12 months per year
    investment_date_periods = 12.0

    #calculate the portion or amount that we plan to invest for each investment date period
    investment_portion = initial_investment/investment_date_periods
    all_investment_dates = pd.date_range(start_date, periods = investment_date_periods, freq = '30d')
    investment_dates = all_investment_dates[all_investment_dates < asset.index[-1]]
    closest_investment_dates = asset.index.searchsorted(investment_dates)

    #get a list of stock prices at the closest investment date
    stock_prices = asset['PTTEP'][closest_investment_dates]

    #get the total number of shares that we invested within 1 year by sum of all of the shares purchased on each of the investment dates
    total_shares_invested = sum(investment_portion/stock_prices)

    uninvested_cash = investment_portion * sum(all_investment_dates >= asset.index[-1])

    total = uninvested_cash + asset['PTTEP'][-1] * total_shares_invested
    return total
```

```python
DCA_List = []
for  date in asset.index:
    DCA_List.append(DCA(date,initial_investment))
```

```python
dca = pd.DataFrame()
dca['DCA'] = DCA_List

#set the date as the index
dca = dca.set_index(pd.DatetimeIndex(asset.index))

#convert the dataframe to a serires
dca = dca.squeeze()
```

```python
#get the votality
lump_sum_vol = lump_sum.pct_change().std()
dca_vol = dca.pct_change().std()

print('Lump Sum Votality:', lump_sum_vol)
print('DCA Votality:', dca_vol)
```

```python
#get difference
difference = dca - lump_sum

#plot the difference
difference.plot(figsize = (16,8), label = 'Difference')
plt.fill_between(difference.index, y1 = difference, y2 = 0, color = 'green', where = difference >0, label = 'DCA > Lump Sum')
plt.fill_between(difference.index, y1 = difference, y2 = 0, color = 'red', where = difference <0, label = 'DCA < Lump Sum')

plt.title('Difference (DCA - Lump Sum)')
plt.legend()
```

จากกราฟแสดงให้เห็นว่า **ถ้าเป็นช่วงราคาหุุ้นขาลงการทำ DCA จะได้ผลตอบแทนดีกว่าการลงทุนเป็น lump sum ในทางกลับกันหากเป็นช่วงราคาขาขึ้นการลงทุนแบบ lump sum จะให้ผลตอบแทนที่ดีกว่า**

```python
lump_sum_win_percentage = round(sum(difference<0)/len(difference) *100,1)
dca_win_percentage = round(sum(difference>0)/len(difference) *100,1)

print('Lump sum Investment Strategy returned more than DCA: ' + str(lump_sum_win_percentage)+ '% of the time.')
print('DCA returned more than lump sum: '+str(dca_win_percentage)+'% of the time.')

```
